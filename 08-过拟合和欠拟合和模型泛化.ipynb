### 过拟合和欠拟合

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(666)
x = np.random.uniform(-3.0, 3.0, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x, y)
plt.show()
```

### 使用线性回归


```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)

# 线性回归默认使用的是R^2 来衡量预测的准确度，这里R方的值是0.49，R方来表示线性关系的强弱
lin_reg.score(X, y)
```

```python
y_predict = lin_reg.predict(X)
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict[np.argsort(x)], color='r')
plt.show()
```

```python
# 使用均方误差来衡量预测的准确度
from sklearn.metrics import mean_squared_error

y_predict = lin_reg.predict(X)
mean_squared_error(y, y_predict)
```

这里发现我们使用线性回归的均方误差如上，接下来和多项式回归来进行预测，来对比均方误差。

### 使用多项式回归

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler

def PolynomialRegression(degree):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)),
        ("std_scaler", StandardScaler()),
        ("lin_reg", LinearRegression())
    ])
```

#### degree = 2

```python
poly2_reg = PolynomialRegression(degree=2)
poly2_reg.fit(X, y)
```

```python
y2_predict = poly2_reg.predict(X)
mean_squared_error(y, y2_predict)
```

这里得到的均方误差是小于使用线性回归的误差的。

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y2_predict[np.argsort(x)], color='r')
plt.show()
```

#### degree = 10

```python
# 将样本对应的次幂升级到10
poly10_reg = PolynomialRegression(degree=10)
poly10_reg.fit(X, y)

y10_predict = poly10_reg.predict(X)
mean_squared_error(y, y10_predict)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y10_predict[np.argsort(x)], color='r')
plt.show()
```

#### degree = 100

```python
# 次幂升到100
poly100_reg = PolynomialRegression(degree=100)
poly100_reg.fit(X, y)

y100_predict = poly100_reg.predict(X)
mean_squared_error(y, y100_predict)
```

```python
均方误差更小了！！！
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y100_predict[np.argsort(x)], color='r')
plt.show()
```

```python
# 这里和前面uniform的区别是，这里是均匀的取值的，两个点之间的距离是一样的
X_plot = np.linspace(-3, 3, 100).reshape(100, 1)
y_plot = poly100_reg.predict(X_plot)
```

```python
plt.scatter(x, y)
plt.plot(X_plot[:,0], y_plot, color='r')
plt.axis([-3, 3, -1, 10])
plt.show() 
# 下面这条曲线是，是degree=100之后真正的曲线。
```

如此，就展示了欠拟合和过拟合。

## 模型泛化：

我们得到了一个模型，当有新的数据来了之后，我们是否能够很好的得到预测值。

我们去衡量我们的模型对于训练数据集的拟合程度有多好，是没有意义的，我们最终需要衡量的是我们得到的模型的泛化能力有多好。

做法就是将数据集分成：训练数据集和测试数据集

### train test split的意义

train test split 能够很好的度量我们得到的模型是过拟合还是欠拟合。

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

#### 线性回归

```python
lin_reg.fit(X_train, y_train)
y_predict = lin_reg.predict(X_test)

# 这里就可以得到该模型的泛化能力，因为预测的数据集是在训练的时候没有遇到的。
mean_squared_error(y_test, y_predict)
```

#### 多项式回归 degree = 2

```python
poly2_reg.fit(X_train, y_train)
y2_predict = poly2_reg.predict(X_test)
mean_squared_error(y_test, y2_predict)
```

#### 多项式回归 degree = 10

```python
poly10_reg.fit(X_train, y_train)
y10_predict = poly10_reg.predict(X_test)
mean_squared_error(y_test, y10_predict)
```

#### 多项式回归 degree = 100

```python
poly100_reg.fit(X_train, y_train)
y100_predict = poly100_reg.predict(X_test)
mean_squared_error(y_test, y100_predict)
```

过拟合：算法所训练的模型过多的表达了数据间的噪音关系。
欠拟合：算法所训练的模型不能完整的表述数据关系。

#### 模型复杂度和模型准确率之间的关系

模型复杂程度越大，训练数据集模型准确率越高；测试数据集先升高后降低。

在网格搜索的时候，对于不同参数训练出来的模型，然后用测试数据集来测试，在测试数据集上表现最好的参数。
